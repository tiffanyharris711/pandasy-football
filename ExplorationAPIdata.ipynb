```python
import requests
import json
from config import api_key
import pandas as pd
import csv
```

```python
# use this url for the API documentation to find the different endpoints
# https://sportsdata.io/developers/api-documentation/nfl

url = "https://api.sportsdata.io/v3/nfl/scores/json/Players?key="
url_response = requests.get(f"{url}{api_key}").json()
```

```python
url_response
```

```python
playerid = []
name = []
height = []
height_ft = [] 
height_in = []
weight = []
experience = [] 
position = []
adp = []
salary = []
    
for i in range(len(url_response)):
    playerid.append(url_response[i]["PlayerID"])
    name.append(url_response[i]["Name"])
    height.append(url_response[i]["Height"])
    height_ft.append(url_response[i]["HeightFeet"])
    height_in.append(url_response[i]["HeightInches"])
    weight.append(url_response[i]["Weight"])
    experience.append(url_response[i]["Experience"])
    position.append(url_response[i]["Position"])
    adp.append(url_response[i]["AverageDraftPosition"])
    salary.append(url_response[i]["UpcomingSalary"])

player_dict = {
    "PlayerID": playerid,
    "Name": name,
    "Height": height,
    "HeightFeet": height_ft,
    "HeightInches": height_in,
    "Weight": weight,
    "Experience": experience,
    "Position": position,
    "ADP": adp,
    "Salary": salary
}

player_df = pd.DataFrame(player_dict)
```

```python
player_df.to_csv("./Resources/player_data.csv")
```
